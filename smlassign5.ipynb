```python

#2) 1. Read the data with pandas and find features and target variables

import pandas as pd
dia_data = pd.read_csv('/content/drive/MyDrive/diabetes_data_upload.csv')
dia_data = dia_data.fillna(0)
print(dia_data.head())
print(dia_data.dtypes)
features = dia_data.drop('class', axis=1)
target = dia_data['class']

```

2

```python
# 2. Plot a graph between features and target

import matplotlib.pyplot as plt
for column in features.columns:
    plt.figure(figsize=(10, 8))
    plt.scatter(features[column], target, color='red')
    plt.title(f'{column} vs Class')
    plt.xlabel(column)
    plt.ylabel('Class')
    plt.grid(True)
    plt.show()
```

```python
# 3. Normalize the data [0 to 1]

norm_data2 = (dia_data - dia_data.min()) / (dia_data.max() - dia_data.min())
print("Normalized Data:")
print(norm_data2)
```

```python
# 4. Convert string valued features to numbers

dia_data = dia_data.replace({'Yes': 1, 'No': 0, 'Positive': 1, 'Negative': 0, 'Male': 1, 'Female': 0})
print("Converted Data with String Features to Numbers:")
print(dia_data)
```
